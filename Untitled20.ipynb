```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# df1 = sns.load_dataset("penguins")
data = pd.read_csv('penguins.csv')
data
```

```python

# Create a Seaborn plot using the loaded data
sns.set(style='whitegrid')  # Set the style of the plot

# Example: Create a scatter plot using Seaborn
sns.scatterplot(x='bill_length_mm', y='flipper_length_mm', data=data)

# Display the plot
plt.show()
```

```python
# Create a Seaborn plot using the loaded data
sns.set(style='whitegrid')  # Set the style of the plot

# Example: Create a scatter plot using Seaborn
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=data)

# Display the plot
plt.show()
```

```python
# # Create a Seaborn plot using the loaded data
# sns.set(style='whitegrid')  # Set the style of the plot

# Example: Create a scatter plot using Seaborn
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', data=data, hue='sex')

# Display the plot
plt.show()
```

```python
# # Create a Seaborn plot using the loaded data
# sns.set(style='whitegrid')  # Set the style of the plot

# Example: Create a scatter plot using Seaborn
sns.lineplot(x='bill_length_mm', y='flipper_length_mm', 
             data=data, hue='sex',
            palette = "rocket_r", markers = ["o", ">"])

plt.grid()
plt.title("this is title")
# Display the plot
plt.show()
```
